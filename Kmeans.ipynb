```
im = itk.imread('/content/drive/My Drive/PIBM_2019_2C/Trabajos Prácticos/TP 1 - Segmentación/Imagen 1.png')
matrix = itk.GetArrayFromImage(im)
(fil,col) = matrix.shape
```

```
cantidadclases = 3
centroides = [[1,2], [100,130], [1,3]]
centroidesnuevos = [[0,0], [0,0], [0,0]]
distanciapixel = np.zeros((256,256))
cant0 = 0
cant1 = 0
cant2 = 0
mean1 = [0,0]
mean2 = [0,0]
mean3 = [0,0]
it = 0

while centroides != centroidenuevos:
  if it != 0:
    centroides = centroidesnuevos
    for i in range(fil):
      for j in range(col):
        distcentr1 = (centroides[0][0] - i)**2 + (centroides[0][1] - j)**2
        distcentr2 = (centroides[1][0] - i)**2 + (centroides[1][1] - j)**2
        distcentr3 = (centroides[2][0] - i)**2 + (centroides[2][1] - j)**2
        lista = [distcentr1, distcentr2, distcentr3]
        index_min = np.argmin(lista)
        distanciapixel[i][j] = index_min

        if distanciapixel[i][j] == 0:
          cant0 = cant0 + 1
          mean1 = mean1 + matrix[i][j]
          
        elif distanciapixel[i][j] == 1:
          cant1 = cant1 + 1
          mean2 = mean2 + matrix[i][j]
          
        else:
          cant2 = cant2 + 1
          mean3 = mean3 + matrix[i][j]
          

  mean1 = mean1/cant0
  mean2 = mean2/cant0
  mean3 = mean3/cant2

  centroidesnuevos = [mean1, mean2 ,mean3]
```
